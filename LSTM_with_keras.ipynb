```python
maxlen = 0
word_freqs = collections.Counter()
num_recs = 0
ftrain = open("./drive/My Drive/Colab Notebooks/RNN_study/training.txt", 'rb')
for line in ftrain:
    label, sentence = line.strip().split(b"t")
    words = nltk.word_tokenize(sentence.decode("ascii", "ignore").lower())
    if len(words) > maxlen:
        maxlen = len(words)
    for word in words:
        word_freqs[word] += 1
    num_recs += 1
ftrain.close()
```
